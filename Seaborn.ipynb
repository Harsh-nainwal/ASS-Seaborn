
Que 1: With Seaborn, you can create various types of plots. Here are five plots along with their uses:

Ans:-
Scatter Plot: Used to visualize the relationship between two continuous variables.
Bar Plot: Useful for comparing categories of data and displaying the distribution of a categorical variable.
Box Plot: Shows the distribution of a continuous variable across different categories and identifies outliers.
Violin Plot: Combines the features of a box plot and a kernel density plot to provide a deeper insight into the distribution of data.
Pair Plot: Displays pairwise relationships between different variables in a dataset, useful for exploring correlations.

```python
#Ques.2
import seaborn as sns

fmri_data = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

```python
#Ques 3
import seaborn as sns

titanic_data = sns.load_dataset("titanic")
sns.boxplot(x="pclass", y="age", data=titanic_data)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
```

```python
#ques 4
import seaborn as sns

diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x="price", hue="cut")
```
